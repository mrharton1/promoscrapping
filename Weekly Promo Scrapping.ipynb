```python
#Scrapping the Fairprice Weekly Promo Page

import scrapy
from scrapy.crawler import CrawlerProcess

class YourSpider(scrapy.Spider):
    name = 'yourspider'
    
    def start_requests(self):
        url = "https://www.fairprice.com.sg/tag/price-drop-buy-now?filter=Country%2FPlace%20Of%20Origin%3ASingapore"
        yield scrapy.Request(url=url, callback=self.parse)
    
    def parse(self, response):
        links = response.css('div.sc-2a85da88-0 > a::attr(href)').extract()  
        for link in links:
            yield response.follow(url=link, callback=self.parse_desc)
    
    def parse_desc(self,response):
        products=response.css('span.sc-aa673588-1.drdope::text').extract()
        prices=response.css('span.sc-aa673588-1.sc-6ac8ef58-5.kQDEta.gbCpHo::text').extract()
        for product, price in zip(products,prices):
            dc[product]=price
    
# Run Spider
dc={}
process = CrawlerProcess()
process.crawl(YourSpider)
process.start()
```
